### Inisiasi Data

```python
import pandas as pd
```

```python
df = pd.read_csv("H05.csv")
```

```python
df
```

### 1. Histogram distribusi nilai matematika

```python
df["nilai_mat"].plot(kind="hist", rwidth=0.8)
```

### 2. Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas

```python
df["fakultas"].value_counts().plot(kind="barh", title="Banyak Mahasiswa Tiap Fakultas", x="Fakultas", y="jumlah")
```

### 3. Diagram pie banyaknya mahasiswa masing-masing fakultas

```python
df["fakultas"].value_counts().plot(kind="pie")
```

### 4. Berdasarkan diagram pie dan batang horizontal, manakah fakultas dengan mahasiswa terbanyak? Manakah diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?

>Berdasarkan diagram pie dan batang horizontal, fakultas dengan mahasiswa terbanyak adalah FTI.

>Menurut saya, diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak adalah diagram batang karena diagram batang sangat baik dalam menampilkan perbedaan mutlak antara jumlah mahasiswa tiap fakultas. Sedangkan diagram pie, menurut saya,  efektif untuk mengilustrasikan hubungan proporsi mahasiswa antar fakultas secara keseluruhan.

### 5. Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y

```python
df.plot(kind="scatter", x="nilai_kim", y="nilai_fis")
```
