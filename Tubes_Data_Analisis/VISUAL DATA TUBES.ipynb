```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\HP\Downloads\datasetTUBES.csv", index_col=0)
```

### VISUALISASI DATA

```python

data = {
    'China': [2017, 14102],
    'Indonesia': [0, 2066]
}
negara_cina_indo = pd.DataFrame(data, index=[1993, 1994])
                     

negara_cina_indo.plot(kind='bar')

plt.xticks(range(len(negara_cina_indo.index)), negara_cina_indo.index)
plt.xticks(rotation=45)

plt.xlabel('Tahun')
plt.ylabel('Jumlah pengguna')
plt.title('Perbedaan Jumlah Pengguna Internet Indonesia dan China')


plt.legend()
plt.show()


```

Jika dibandingkan kedua negara ini memiliki jumlah pengguna yang berbedaa pada tahun 1994 dan 1993  pada tahun 1993 Indonesia pengguna internetnya masih 0 sementara negara china sudah mulai ada pengguna internetnya dari hal tersebut dapat diambil kesimpulan bahwa perkembangan internet di china lebih cepat dari pada di Indonesia.

```python
region=df.loc[(df["Entity"]=="East Asia and Pacific") | (df["Entity"]=="Sub-Saharan Africa") ].loc[df["Year"]==2020]
region.plot(kind="bar",x="Entity", y=["Internet Users(%)","Cellular Subscription","Broadband Subscription"], title="Jumlah pengguna internet pada tahun 2020")
plt.xticks(rotation=45)
```

Dari diagram ini dapat kita lihat bahwa perbedaan region juga memengaruhi banyaknya pengguna internet. Dari tabel itu dapat dilihat ketimpangan yang sanngat besar diantara wilayah East Asia Pasifik dengan Sub-Saharan Afrika Ketimpangan tersebut bisa saja ada dikarenakan perbedaaan perkembangan di kedua wilayah tersebut. Penyebab ketimpangan tersebut dipengaruhi beberapa hal misalnya perbedaan keadaan ekonomi kedua wilayah tersebut dan juga penyebab dari sedikitnya pengguna internet di Sub-Sahara Afrika dikarenakan keterbatasan Infrastruktur.


```python
df2=df.loc[df["Continent"]!="Undefined" ]
continent=df2.loc[ df["Year"]==2020].groupby(["Continent"])["No. of Internet Users"].agg('sum').reset_index()
continent1=df2.loc[ df["Year"]==2015].groupby(["Continent"])["No. of Internet Users"].agg('sum').reset_index()
colls=continent1['No. of Internet Users']
continent=pd.concat([colls ,continent], axis=1)
continent = continent.rename(columns={'No. of Internet Users': '2015'})
continent.columns.values[2] = '2020'
continent.set_index('Continent')

continent.plot(kind='bar')
plt.xticks(range(len(continent.index)), continent["Continent"])
plt.xticks(rotation=45)


plt.xlabel('Benua')
plt.ylabel('Jumlah pengguna')
plt.title('Perbedaan Jumlah Pengguna Internet Indonesia dan China')


plt.legend()
plt.show()
```

Berdasarkan tabel diatas dapat dilihat bahwa jika dibandingkan dengan benua lain, benua Asia memiliki jumlah pengguna internet terbanyak hal ini terjadi karena jika dibandingkan dengan benua lain benua Asia memiliki wilayah yang jauh lebih besar hal ini menyebabkan pengguna internet di asia menjadi yang paling banyak diantar benua lainnya. selain itu dari tabel tersebut kita juga dapat mengetahui untuk benua asia pengguna internet di tahun 2015 dan 2020 terdapat ketimpangan yang besar, arti dari ketimpangan tersebut adalah perkembangan teknologi di benua Asia termasuk cepat dan merata sehingga terdapat perbedaan yang besar diantara kedua tahun tersebut.
