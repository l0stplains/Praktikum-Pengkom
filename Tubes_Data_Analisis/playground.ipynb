```python
df['Continent'] = df['Entity']
```

```python
import pycountry_convert as pc
```

```python
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
```

```python
for i, country in enumerate(df['Continent']):
  try:
    df['Continent'][i] = country_to_continent(country)
  except:
    df['Continent'][i] = "Undefined"
```

```python
country_to_continent("Indonesia")
```

```python
with pd.ExcelWriter('output.xlsx') as writer:  
    df.to_excel(writer)
```

```python
list_negara_galengkap = df.groupby("Entity").size() != 41
```

```python
list_negara_galengkap_beneran = []
for i in range(len(df['Entity'].unique())):
  if(list_negara_galengkap[i] == False):
    list_negara_galengkap_beneran.append(df['Entity'].unique()[i])
```

```python
df = df[df['Entity'].isin(list_negara_galengkap_beneran)]
```

```python
with pd.ExcelWriter('fix.xlsx') as writer:  
    df.to_excel(writer)
```
