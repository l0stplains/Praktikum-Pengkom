```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
df.describe()
```

```python
#Data ini berisi jumlah pengguna internet, persentase pengguna internet, jumlah pengguna jaringan seluler, dan jumlah pengguna jaringan broadband di berbagai negara per tahun
#Cara mengakses dan mengetahui berbagai informasi terkait data bisa dilakukan dengan dua kode diatas
```

```python
nama_file = 'dataset.csv'
ukuran_bytes = os.path.getsize(nama_file)

ukuran_kb = ukuran_bytes / 1024
ukuran_mb = ukuran_kb / 1024

ekstensi_file = os.path.splitext(nama_file)[1]

print(f"Format file: {ekstensi_file}")
print(df.shape)
print(f"Ukuran file {nama_file}: {ukuran_bytes} bytes, {ukuran_kb:.2f} KB, {ukuran_mb:.2f} MB")

```

```python
#Dengan kode di atas,terlihat bahwa data ini berformat csv dengan dimensi 6888x9 dan berukuran 427.3 KB
#Data ini didapat dari https://www.kaggle.com/datasets/ashishraut64/interner-users/data
```

```python
entities = ["Indonesia", "United Kingdom","Japan","Germany","Hong Kong","Canada"]
df2 = df[df["Entity"].isin(entities)]

# Plotting grafik garis dengan warna yang berbeda
plt.figure(figsize=(10, 6))

for entity in entities:
    entity_data = df2[df2["Entity"] == entity]
    plt.plot(entity_data["Year"], entity_data["No. of Internet Users"], label=entity, marker='o')


# Menambahkan judul dan label sumbu
plt.title('Jumlah Pengguna Internet Terhadap Waktu')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pengguna Internet (dalam juta)')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```

```python
entities = ["World", "Indonesia", "United Kingdom","Japan","Germany","Hong Kong","Canada"]
df2 = df[df["Entity"].isin(entities)]

# Plotting grafik garis dengan warna yang berbeda
plt.figure(figsize=(10, 6))

for entity in entities:
    entity_data = df2[df2["Entity"] == entity]
    plt.plot(entity_data["Year"], entity_data["Internet Users(%)"], label=entity, marker='o')


# Menambahkan judul dan label sumbu
plt.title('Persentase Pengguna Internet Terhadap Waktu')
plt.xlabel('Tahun')
plt.ylabel('Pengguna Internet (%)')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```

```python
#Dari visualisasi diatas terlihat bahwa:
#Pengguna Internet di Indonesia jauh lebih banyak daripada negara-negara maju
#Akan tetapi, jika dilihat dari persentase terhadap jumlah penduduk, Indonesia masih tertinggal jauh
#Persentase pengguna internet di negara-negara maju lebih dari 80 persen, sementara Indonesia kurang dari 60 persen
#Bahkan, persentase pengguna internet di Indonesia di tahun 2020 masih lebih rendah daripada persentase dunia
#Persentase pengguna internet di negara maju meningkat pesat di pertengahan tahun 1990an,sementara di Indonesia persentase pengguna internet masih dibawah 5 persen

```

```python
entities = ["Indonesia", "United Kingdom","Japan","Germany","Hong Kong","Canada"]
df2 = df[df["Entity"].isin(entities)]

# Plotting grafik garis 
plt.figure(figsize=(10, 6))

for entity in entities:
    entity_data = df2[df2["Entity"] == entity]
    plt.plot(entity_data["Year"], entity_data["Cellular Subscription"], label=entity, marker='o')


# Menambahkan judul dan label sumbu
plt.title('Jumlah Pengguna Jaringan Seluler Terhadap Waktu')
plt.xlabel('Tahun')
plt.ylabel('Pengguna Jaringan Seluler (dalam juta)')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```

```python
entities = ["Indonesia", "United Kingdom","Japan","Germany","Hong Kong","Canada"]
df2 = df[df["Entity"].isin(entities)]

# Plotting grafik garis 
plt.figure(figsize=(10, 6))

for entity in entities:
    entity_data = df2[df2["Entity"] == entity]
    plt.plot(entity_data["Year"], entity_data["Broadband Subscription"], label=entity, marker='o')


# Menambahkan judul dan label sumbu
plt.title('Jumlah Pengguna Jaringan Broadband Terhadap Waktu')
plt.xlabel('Tahun')
plt.ylabel('Pengguna Jaringan Broadband (dalam juta)')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```
