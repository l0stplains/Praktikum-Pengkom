### STATISTIK

```python
a=df.loc[df["Entity"]=="World"]
a.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9]).drop(columns='Year')
```

```python
countries_2020= df[(df["Continent"]!="Undefined") &(df["Year"]==2020) & (df["No. of Internet Users"]!=0)] 
countries_2020.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9]).drop(columns='Year')
```

```python

buru=df.iloc[countries_2020["Internet Users(%)"].idxmin()]
bahri=df.iloc[countries_2020["Internet Users(%)"].idxmax()]
result=pd.concat([buru,bahri], axis=1)
result = result.rename(columns={1024: 'MIN', 450: 'MAX'}).drop('Unnamed: 0',axis=0)
result
```

Berdasarkan data yang di atas, perbedaan persentase pengguna internet antara Burundi dan Bahrain sangat signifikan.

1.  **Burundi:** Persentase pengguna internet hanya sekitar 9.4%. Angka ini menunjukkan bahwa jumlah penduduk yang menggunakan internet di negara ini relatif rendah pada tahun 2020.
    
2.  **Bahrain:** Sementara di Bahrain, persentase pengguna internet sangat tinggi, mencapai sekitar 99.67%. Angka ini menandakan bahwa hampir semua penduduk di Bahrain telah mengakses internet pada tahun tersebut.
    

Perbedaan ini menunjukkan kesenjangan yang besar dalam akses dan penetrasi internet antara kedua negara tersebut. Bahrain memiliki tingkat penetrasi internet yang hampir mendekati keseluruhan populasi, sementara Burundi masih memiliki tingkat akses yang jauh lebih rendah.

Kesenjangan seperti ini dapat memiliki dampak besar terutama dalam hal akses informasi, komunikasi, kesempatan pendidikan, dan pertumbuhan ekonomi karena akses yang terbatas terhadap internet dapat membatasi akses terhadap informasi dan kesempatan lainnya.

Untuk meningkatkan persentase pengguna internet di Burundi, pemerintah dapat fokus pada beberapa langkah strategis. Investasi dalam infrastruktur telekomunikasi yang lebih luas dan stabil diperlukan, terutama di daerah pedesaan. Kebijakan pro-internet serta insentif untuk perusahaan telekomunikasi bisa mendorong pertumbuhan ekosistem digital. Program pendidikan dan literasi digital akan memperkuat pemahaman tentang manfaat internet. Kolaborasi dengan organisasi internasional dan swasta, serta memastikan akses internet terjangkau, juga akan menjadi kunci dalam memperluas akses internet bagi penduduk. Inovasi teknologi lokal juga dapat membantu menciptakan solusi yang lebih terjangkau dan efektif untuk meningkatkan akses internet di negara tersebut.
